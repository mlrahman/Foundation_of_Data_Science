```python
import numpy as np
import csv

def matrix_normalization(x):
    min_value=x.min()
    max_value=x.max()
    min_range=-1
    max_range=1
    x=min_range+(((x-min_value)*(max_range-min_range))/(max_value-min_value))
    return x


arr = np.empty((0,10), dtype=float)

with open('hcvdat0.csv', mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    
    fl=1;
    for row in csv_reader:
        if fl == 1: 
            fl=2
            continue
        if 'NA' not in row[4:14]:
            arr = np.append(arr, np.array([row[4:14]],dtype=float), axis=0)
        
y=matrix_normalization(arr)


for data in y:
    print(data)
```
