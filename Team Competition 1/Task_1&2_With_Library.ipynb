```python
import numpy as np
import csv
from sklearn import preprocessing


def matrix_normalization(x):
    minmax_scale = preprocessing.MinMaxScaler(feature_range=(-1, 1))
    x = minmax_scale.fit_transform(x)
    return x



arr = np.empty((0,10), dtype=float)

with open('hcvdat0.csv', mode='r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    
    fl=1;
    for row in csv_reader:
        if fl == 1: 
            fl=2
            continue
        if 'NA' not in row[4:14]:
            arr = np.append(arr, np.array([row[4:14]],dtype=float), axis=0)

            
y=matrix_normalization(arr)


for data in y:
    print(data)
```
